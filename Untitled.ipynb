```python
# import necessary lib for analysis
import json
import pandas as pd
from rich import print

# load file to be analyzed
file= 'stats.json'
data= None

with open(file, 'r') as f:
    data= json.load(f)

channl_id, stats= data.popitem() #returns a tuple of both key and values of the dict
channl_stats= stats['channel_statistics'] #get channel_stats
video_stats= stats['video_data'] #get video stats
 
```

```python
# channel statisticts
print(f'views: {channl_stats['viewCount']}')
print(f'subscriber: {channl_stats['subscriberCount']}')
print(f'videos: {channl_stats['videoCount']}')
```

```python
# video statistics
# Here we will sort the video data according to the vids stats i.e viewcount, the vids with the most viewed will be on top sorted down to the least

sorted_vids= sorted(video_stats.items(), key= lambda item: int(item[1]['statistics']['viewCount']), reverse=True) #to achieve this we use the built in 'sorted func'
stats= []
for vid in sorted_vids:
    vid_id= vid[0]
    title= vid[1]['snippet']['title']
    views= int(vid[1]['statistics'].get('viewCount'))
    likes= int(vid[1]['statistics'].get('likeCount'))
    comments= int(vid[1]['statistics'].get('commentCount'))
    stats.append([title, views, likes, comments])
```

```python
# put the vids stats into a pandas dataframe
df= pd.DataFrame(stats, columns=['title', 'views', 'likes', 'comments'])
df.head(10)# get the first 10 rows of the table
```

```python
# get the first top 10 data and plot it using pandas
top10= df.head(10)
ax= top10.plot.bar(x='title', y='views', figsize=(12, 8), fontsize=14)
```

```python
# get the first bottom 10 data and plot it using pandas
bottom10= df.tail(10)
ax= bottom10.plot.bar(x='title', y='views', figsize=(12, 8), fontsize=14)
```

```python
# analyze stats for top 10 likes/comments avg_ratio
likes_avg= top10['likes'].mean()
comments_avg= top10['comments'].mean()
print('Top 10')
print('avg_likes:', likes_avg)
print('avg_comments:', comments_avg)

# analyze stats for bottom 10 likes/comments avg_ratio
likes_avg= bottom10['likes'].mean()
comments_avg= bottom10['comments'].mean()
print('Bottom 10')
print('avg_likes:', likes_avg)
print('avg_comments:', comments_avg)

```
